```python
import numpy as np
from scipy.optimize import linear_sum_assignment


def find_optimal_P(M):
    # 주어진 행렬 M의 크기를 얻음
    n = M.shape[0]

    # linear_sum_assignment 함수를 사용하여 P를 찾음
    row_ind, col_ind = linear_sum_assignment(-M)

    # P를 생성
    P = pd.DataFrame(np.zeros_like(M), index=M.index, columns=M.columns)
    for i in range(len(row_ind)):
        P.iloc[row_ind[i], col_ind[i]] = 1

    return P
```
