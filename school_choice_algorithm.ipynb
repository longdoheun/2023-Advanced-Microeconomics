```python
class Student:
  def __init__(self, name=str, preference_list=list):
    self.name = name
    self.preference_list = preference_list

  def apply(self):
    '''
    preference list means their each preference school list, which the smaller its index is, the more preferred one is.
    '''
    school_to_apply = self.preference_list.pop(0)

    return school_to_apply
```

```python
class School:
  def __init__(self, name=str, priority_list=list, capacity=int):
    self.name = name
    self.priority_list = priority_list
    self.capacity = capacity
    self.current_accepted_list = [] # initial list is empty

  def sort_by_priority(self, applied_list=list):
    '''
    before accepting, check the prioriy of possible choices by sorting algorithms
    '''
    list_to_sort = self.current_accepted_list + applied_list # priority check between new one and existing one

    for i in range(len(list_to_sort)):
      for j in range(len(list_to_sort)):
        if self.priority_list.index(list_to_sort[i]) < self.priority_list.index(list_to_sort[j]):
          list_to_sort[i], list_to_sort[j] = list_to_sort[j], list_to_sort[i]
        else:
          continue

    list_sorted = list_to_sort
      
    return list_sorted

  def check_acceptance(self, applied_list=list):
    '''
    Accept individuals as long as it has a capacity to accept
    '''
    prioritized_accepted_list = self.sort_by_priority(applied_list)
    self.current_accepted_list = prioritized_accepted_list[:self.capacity]
    
    return self.current_accepted_list # just in case, current acceptance
```

```python
def Deferred_Acceptace(I=list, S=list):
  '''
  applied list : the applied list of individuals in that round
  currently accepted : only individual not currently accepted can apply to most preferred school in that round
  q : set of capacity of schools
  '''
  iteration = 0
  currently_accepted = []
  q = { school.name : school.capacity for school in S}
  NAME_OF_I = [individual.name for individual in I]

  while True:
    iteration += 1
    applied_list = { school.name : [] for school in S}
    print(f"\n{iteration} Round Appliance")
    print(f"currently_accepted : {currently_accepted}")

    if set(currently_accepted) == set(NAME_OF_I) :
      print(f"DA finished")
      result = { school.name : school.current_accepted_list for school in S}
      return result
    
    else:
      ### apply to most preferred school in that round
      for individual in I:
        if individual.name not in currently_accepted:
          school_to_apply = individual.apply()
          applied_list[school_to_apply].append(individual.name)
          print(f"{individual.name} applied to {school_to_apply}")
      
      print(f"applied_list : {applied_list}")

      currently_accepted = [] # initialize accepted list before accepting

      ### after applying, check the capacity of each school and reject
      for school in S:
        if len(applied_list[school.name]+school.current_accepted_list) <= q[school.name]:
          school.current_accepted_list += applied_list[school.name]
        else:
          school.check_acceptance(applied_list[school.name])
        
        currently_accepted += school.current_accepted_list # reflect new acceptance and rejection
        print(f"{school.name} accepted {school.current_accepted_list} ")
```

```python
i1 = Student("i1", ["s1","s2","s3"])
i2 = Student("i2", ["s1","s2","s3"])
i3 = Student("i3", ["s1","s2","s3"])
i4 = Student("i4", ["s2","s1","s3"])
i5 = Student("i5", ["s2","s1","s3"])

s1 = School("s1", ["i1", "i4", "i2", "i3", "i5"], 2)
s2 = School("s2", ["i5", "i2", "i3", "i4", "i1"], 2)
s3 = School("s3", ["i1", "i2", "i3", "i4", "i5"], 1)
```

```python
I = [i1, i2, i3, i4, i5] # set of individual
S = [s1, s2, s3] # set of schools

Deferred_Acceptace(I, S)
```
