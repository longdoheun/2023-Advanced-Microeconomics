```python
import pandas as pd
import numpy as np
import random
```

```python
df = pd.read_excel("./oh_preferred_time.xlsx", index_col=0)
df
```

각 element의 '컴마로 구분된 이름들'을 '이름들의 list'로 변환

```python
for index, row in df.iterrows():
    for day in df.columns:
      comma_separated_value = row[day].replace(" ", "")
      value_array = comma_separated_value.split(",")
      df.loc[index, day] = value_array
```

```python
def find_min_index(df):
  len_df_values = df.values # 데이터프레임을 넘파이 배열로 변환
  try:
    min_length = np.min(len_df_values[len_df_values > 0]) # 0보다 큰 요소들 중 최솟값 구하기
  except:
    return False

  for index, row in df.iterrows():
      for day in df.columns:
        if row[day] == min_length:
          return (day, index)
```

```python
def get_len_df(df):
  copied_df = df.copy()
  for index, row in copied_df.iterrows():
    for day in copied_df.columns:
      copied_df.loc[index, day] = len(row[day])
  return copied_df
```

greedy algorithm 이기 때문에 결과가 시행할 때마다 바뀔 수 있음.

```python
temp_df = df.copy() # init
allocated = [] # 배정된 인원을 담은 리스트

while True:
  len_df = get_len_df(temp_df)

  if find_min_index(len_df):
    day, index = find_min_index(len_df)
    min_element = temp_df[day][index].copy()
    random_index = random.randint(0, len(min_element) - 1)  # 무작위 인덱스 선택
    popped_element = min_element.pop(random_index)
    print(day, index, min_element, popped_element)
    allocated.append(popped_element)
    temp_df[day][index] = []
    df[day][index] = popped_element

    for index, row in temp_df.iterrows():
      for day in temp_df.columns:
        temp_list = temp_df.loc[index, day].copy()
        if popped_element in temp_list:
          temp_list.remove(popped_element)
          temp_df.loc[index, day] = temp_list
    
  else:
    current = temp_df
    break

current
```
